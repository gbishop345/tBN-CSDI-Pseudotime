# Run CSDI on reordered RNA data

Runs `exe_rna_reorder.py`. **Pseudotime method** (`--pseudotime-method`): **`dpt`** → `data/rna/rna_reordered_dpt.csv`, `save/rna_reorder_<time>/`; **`slingshot`** → `rna_reordered_slingshot.csv`, `save/rna_reorder_slingshot_<time>/`; **`phate`** → `rna_reordered_phate.csv`, `save/rna_reorder_phate_<time>/`. Build the CSV with `python reorder_rna.py --dataset rna --method dpt|slingshot|phate` (Slingshot: `rpy2` + R `slingshot`; PHATE: `pip install phate`).

By default the script sweeps missing ratios **0.1, 0.3, 0.5, 0.7, 0.9** with **5 trials** each. Outputs include **`trials_summary.json`** and **`summary_mean_std.json`**. Same pattern for `exe_rna.py` → `save/rna_<time>/`.

```python
# Optional: override defaults (edit as needed)
CONFIG = "base.yaml"
DEVICE = "cuda:0"
SEED = 1
NSAMPLE = 100
NTRIALS = 5  # trials per missing ratio
PSEUDOTIME_METHOD = "dpt"  # "dpt" | "slingshot" | "phate" (must match reorder_rna.py --dataset rna --method)
# Default script behavior: missing ratios .1,.3,.5,.7,.9 each NTRIALS times.
# Set SINGLE_MISSING_RATIO to a float (e.g. 0.5) to run only that ratio; leave None for full sweep.
SINGLE_MISSING_RATIO = None
MODELFOLDER = ""  # e.g. "rna_reorder_20250101_120000" to load checkpoint instead of training
```

```python
import subprocess
import sys

cmd = [
    sys.executable, "exe_rna_reorder.py",
    "--pseudotime-method", PSEUDOTIME_METHOD,
    "--config", CONFIG,
    "--device", DEVICE,
    "--seed", str(SEED),
    "--nsample", str(NSAMPLE),
    "--ntrials", str(NTRIALS),
]
if SINGLE_MISSING_RATIO is not None:
    cmd += ["--testmissingratio", str(SINGLE_MISSING_RATIO)]
if MODELFOLDER:
    cmd += ["--modelfolder", MODELFOLDER]

subprocess.run(cmd)
```
